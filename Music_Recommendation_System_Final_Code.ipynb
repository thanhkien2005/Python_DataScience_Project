# **I. Giới thiệu**

Hệ thống gợi ý âm nhạc là một công nghệ sử dụng các thuật toán và mô hình học máy để cung cấp các đề xuất âm nhạc phù hợp với sở thích của người dùng. Với sự phát triển nhanh chóng của nền tảng âm nhạc trực tuyến, nhu cầu về các giải pháp giúp người dùng tìm kiếm và khám phá âm nhạc mới ngày càng tăng.

### Tại sao cần hệ thống gợi ý âm nhạc?

1. **Khám Phá Âm Nhạc Mới**: Người dùng thường gặp khó khăn trong việc tìm kiếm các bản nhạc mới phù hợp với phong cách và sở thích cá nhân. Hệ thống gợi ý giúp họ dễ dàng khám phá âm nhạc mới.

2. **Tăng Trải Nghiệm Người Dùng**: Một hệ thống gợi ý tốt không chỉ giúp người dùng tìm thấy âm nhạc mà còn cải thiện trải nghiệm nghe nhạc tổng thể.

3. **Tối Ưu Hoá Nội Dung**: Bằng cách phân tích dữ liệu người dùng, hệ thống có thể tối ưu hóa nội dung mà người dùng thấy, từ đó tăng cường mức độ tương tác.

### Các Phương Pháp Thông Dụng

- **Lọc Nội Dung (Content-Based Filtering)**: Dựa trên các đặc trưng của bài hát (như thể loại, nghệ sĩ, tempo) để gợi ý các bài hát tương tự.

- **Lọc Hợp Tác (Collaborative Filtering)**: Sử dụng dữ liệu từ nhiều người dùng để tìm ra các mối liên hệ và sở thích tương tự, từ đó đưa ra các đề xuất.

- **Học Sâu (Deep Learning)**: Áp dụng các mô hình học sâu để phát hiện các mẫu phức tạp trong dữ liệu âm nhạc và hành vi người dùng.

Hệ thống gợi ý âm nhạc không chỉ giúp người dùng tiết kiệm thời gian mà còn mở ra những trải nghiệm âm nhạc phong phú và đa dạng hơn.

```python
# boxplot for numerical columns
sns.set_style('darkgrid')
sns.set(rc={"axes.facecolor":"#F2EAC5","figure.facecolor":"#F2EAC5"})
columns = ['popularity', 'duration_ms', 'tempo', 'loudness', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence']
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 10))
for i, col in enumerate(columns):
    sns.boxplot(y=col, data=numerical_cols, ax=axes[i//4, i%4])
    axes[i//4, i%4].set_title(col)
plt.tight_layout()
plt.show()
```

  - Genres of Top Artists: Các thể loại được thể hiện trong bài hát của nghệ sĩ hàng đầu







##### i. Top Genres

Mức độ phổ biến của các thể loại được xác định bởi mức độ phổ biến trung bình của các bài hát trong mỗi thể loại. Sử dụng hàm groupby() để nhóm dữ liệu theo thể loại và tính toán mức độ phổ biến trung bình của các bài hát trong từng thể loại.



```python
top_genres = df.groupby('genre')['popularity'].mean()
top_genres = top_genres[top_genres >= 35].sort_values(ascending=False)
top_genres[:10]
```

##### ii. Top Artists

Cột nghệ sĩ chứa danh sách nghệ sĩ của mỗi bài hát, cách nhau bằng dấu chấm phẩy. Để trích xuất các nghệ sĩ, ta làm như sau:
  - Chia các nghệ sĩ ở mỗi hàng thành danh sách nghệ sĩ bằng hàm split().
  - Sử dụng hàm explode() để tạo một hàng mới cho mỗi nghệ sĩ trong danh sách



```python
copy_df = df.copy()
copy_df['artist_name'] = copy_df['artist_name'].str.split(';')
copy_df = copy_df.explode('artist_name')
```

Giờ đây, ta có thể nhóm dữ liệu theo nghệ sĩ và tính toán mức độ phổ biến trung bình của các bài hát theo từng nghệ sĩ để xác định những nghệ sĩ hàng đầu.


```python
top_artists = copy_df.groupby('artist_name')['popularity'].mean()
top_artists = top_artists[top_artists >= 70].sort_values(ascending=False)
top_artists[:10]
```

##### iii. Genres of top Artists

  - Sau khi xác định được các nghệ sĩ hàng đầu, sử dụng hàm isin() để lọc dữ liệu để chỉ bao gồm các bài hát của các nghệ sĩ hàng đầu.
  - Sau đó, sử dụng hàm unique() để trích xuất các thể loại unique() được thể hiện trong bài hát của các nghệ sĩ hàng đầu.


```python
top_artists = top_artists.index.tolist()
genres_artists = df[df['artist_name'].isin(top_artists)]['genre'].unique().tolist()
genres_artists[:10]
```

##### i. Phân phối thể loại - Distribution of number of genres

 **Sử dụng hàm groupby() để nhóm dữ liệu theo nghệ sĩ và đếm số lượng thể loại cho tất cả nghệ sĩ và nghệ sĩ hàng đầu.**

```python
copy_df = df.copy()
copy_df['artist_name'] = copy_df['artist_name'].str.split(';')
copy_df = copy_df.explode('artist_name')

artist_genre_count = copy_df.groupby('artist_name')['genre'].nunique().sort_values(ascending=False)

top_artists = copy_df.groupby('artist_name')['popularity'].mean()
top_artists = top_artists[top_artists >= 60].sort_values(ascending=False)
top_artists = top_artists.index.tolist()

copy_df = copy_df[copy_df['artist_name'].isin(top_artists)]

top_artist_genre_count = copy_df.groupby('artist_name')['genre'].nunique().sort_values(ascending=False)
```

**Sau đó, ta tạo 2 biểu đồ để trực quan hóa sự phân bổ số lượng thể loại cho tất cả nghệ sĩ và nghệ sĩ hàng đầu.**


```python
fig, axes = plt.subplots(1, 2, figsize=(20, 6))
sns.set_style('darkgrid')


axes[0].hist(artist_genre_count, bins=20)
axes[0].set_title("Distribution of Genres per Artist")
axes[0].set_xlabel("Number of Genres")
axes[0].set_ylabel("Number of Artists")
for bar in axes[0].containers[0]:
    height = bar.get_height()
    if height > 0:
        axes[0].text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom', fontsize=10)
axes[0].set_xticks(np.arange(0, 21, 1))


axes[1].hist(top_artist_genre_count, bins=13)
axes[1].set_title("Distribution of Genres per Top Artist")
axes[1].set_xlabel("Number of Genres")
axes[1].set_ylabel("Number of Artists")
for bar in axes[1].containers[0]:
    height = bar.get_height()
    if height > 0:
        axes[1].text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom', fontsize=10)
axes[1].set_xticks(np.arange(0, 14, 1))

plt.show()
```

```python
song_cluster_pipeline = Pipeline([('scaler', StandardScaler()),
                                  ('kmeans', KMeans(n_clusters=30,
                                   verbose=False))
                                 ], verbose=False)

X = df.select_dtypes(np.number)
song_cluster_pipeline = song_cluster_pipeline.fit(X)
```

```python
def get_song_data(song, spotify_data):
    try:
        song_data = spotify_data[(spotify_data['track_name'] == song['name']) &
                                 (spotify_data['year'] == song['year'])].iloc[0]
        return song_data
    except IndexError:
        return None



def get_mean_vector(song_list, spotify_data):
    song_vectors = []
    for song in song_list:
        song_data = get_song_data(song, spotify_data)
        if song_data is None:
            print('Warning: {} does not exist in Spotify or in database'.format(song['name']))
            continue
        song_vector = song_data[numerical_cols].values
        song_vectors.append(song_vector)

    song_matrix = np.array(list(song_vectors))
    return np.mean(song_matrix, axis=0)

def recommend_songs(song_list, spotify_data, n_songs=10):
    numerical_cols = song_cluster_pipeline.steps[0][1].feature_names_in_
    spotify_data_aligned = spotify_data[numerical_cols]

    # Tính vector trung tâm từ các bài hát đầu vào
    song_center = get_mean_vector(song_list, spotify_data)
    scaler = song_cluster_pipeline.steps[0][1]
    scaled_data = scaler.transform(spotify_data_aligned)
    song_center_scaled = scaler.transform(song_center.reshape(1, -1))

    # Tính khoảng cách và chọn các bài hát gần nhất
    distances = np.linalg.norm(scaled_data - song_center_scaled, axis=1)
    recommendations = spotify_data.iloc[np.argsort(distances)[:n_songs]]

    # Định dạng kết quả
    formatted_recommendations = []
    for _, row in recommendations.iterrows():
        song_entry = f"{row['track_name']} - {row['year']} - {row['artist_name']}"
        formatted_recommendations.append(song_entry)

    return formatted_recommendations

```
